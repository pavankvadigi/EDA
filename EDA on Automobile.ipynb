# EDA on Automobile dataset

### 1. What is EDA?

__EDA__ or __Exploratory Data Analysis__ is nothing but a data exploration technique to understand the various aspects of the data.

__EDA__ is the first step and critical in analyzing the data. As part of __EDA__, the following list of things are involved.

    * Understand the data
    * Data Cleaning
    * Filter the data from redundancies
    * Finding the patterns
    * Identify the relationships between variables
    * Set of transformations
    * Deriving simple models
    * Conclusion

### Data Cleaning / Pre Processing

The output from the value_counts represents that __"?"__ mark symbol is stored in the price attribute. When no data or invalid data is stored for a feature in a particular observation, we can say this feature has missing values.

Examining this is important because when some of our data is missing, it can lead to weak or biased analysis.

Let's convert the __"?"__ symbol to __"null"__ value with the help of replace function.

#### Replacing the "?" mark symbol with null values in the price column using np.nan method

```python
automobile_data.price = automobile_data.price.replace("?",np.nan) 
```

To validate whether the previous statement has converted the ? mark to Null values

we can handle missing values in many ways.

__Delete:__ We can delete the rows with missing values or delete the whole column which has missing values. The dropna() method from pandas library can be used to accomplish this task.

__Impute:__ Deleting data is the simplest way but it might cause huge amount of information loss. So, replacing data might be a better option than deleting.One of the standard replacement technique is to replace the missing values with statistical methods(Mean/Median/Mode).
For example, we can replace the "bore" column with the mean value of bore column.

### Data Transformation

In our dataset, we have two attributes named __city-mpg__ and __highway-mpg__ indicates car fuel consumption in __miles per gallon__ unit. However, we don't use such mpg, we want to convert the values to Litre/100km.

**Formula to convert from miles per gallon to litres per 100km. 

__Litre/ 100km = 235.21/ x mpg__

```python
auto['city-mpg'] = 235.21/auto['city-mpg']
```

Renaming the mileage columns from mpg to L/100km

```python
auto.rename(columns={'city-mpg':'city-L/100km'},inplace=True)
```

```python
auto['highway-mpg'] = 235.21/auto['highway-mpg']
```

```python
auto.rename(columns={'highway-mpg':'highway-L/100km'},inplace=True)
```

### Symboling
Cars are initially assigned a risk factor symbol associated with its price from an insurer's perspective. 

Symboling value shows how risky or safe a vehicle is? In our dataset the range of __Symboling__ attribute lies between -2 and +3. A value of +3 indicates that the auto is more risky and -2 is pretty safe.

Based on the Symboling value let's divide the data into two categories __"Risky"__ and __"Safe"__

__Note: -3 to 0 indicates a "Safe" vehicle & 1 to 3 indicates the vehicle as "Risky"__

```python
def symboling(x): # Defining a function to describe the vehicle is risky or safe
    if (x>0):
        return "Risky"
    return "Safe"
# Adding a new column named "risk-factor" to categorize the vehicle is risky or safe
auto['risk-factor'] = auto['symboling'].apply(lambda x: symboling(x))
```
